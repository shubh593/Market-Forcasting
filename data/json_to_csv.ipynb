```python
import json
import pandas as pd

with open("market_history.json") as f:
    data = json.load(f)

rows = []

for entry in data:
    result = entry["data"]["result"]
    rows.append({
        "timestamp": entry["fetched_at"],
        "symbol": result.get("symbol"),
        "volume": float(result.get("volume", 0)),
        "high": float(result.get("high", 0)),
        "low": float(result.get("low", 0))
    })

df = pd.DataFrame(rows)
df.to_csv("market_history.csv", index=False)

print("CSV ready for analysis")
```
